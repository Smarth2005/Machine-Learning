```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Observations:
1. When rating is between 4.5 to 4.9 ---- Excellent
2. When rating is between 4.0 to 3.4 ---- Very Good
3. When rating is between 3.5 to 3.9 ---- Good
4. When rating is between 3.0 to 3.4 ---- Average
5. When rating is between 2.5 to 2.9 ---- Average
6. When rating is between 2.0 to 2.4 ---- Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings, palette='rainbow' )
```

```python
# Create figure and axes
fig, ax = plt.subplots()

# Set cream background
fig.patch.set_facecolor('#FFFDD0')  # Figure background
ax.set_facecolor('#FFFDD0')         # Axes background

dark_green = sns.color_palette("dark")[2]
colors = ['white','red','orange','yellow','green', dark_green]
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings, hue='Rating color', palette = colors )
```

Observations:
1. `Not Rated` count is very high.
2. Maximum number of ratings are between 2.5 and 3.4.

```python
## Count Plot

fig, ax = plt.subplots()
fig.patch.set_facecolor('#FFFDD0')
ax.set_facecolor('#FFFDD0')

sns.countplot(x = 'Rating color', data = ratings, palette = colors)
```

```python
## Find countries where restaurants offer table booking, and get the count for each country.
final_df[final_df['Has Table booking']=='Yes'].groupby('Country').size().reset_index(name='Restaurant Count')
```

```python
## Total Restaurants in different countries
final_df.groupby('Country')['Restaurant ID'].nunique().reset_index(name='Unique Restaurant Count')
```

```python
## Which cuisines are most popular across the dataset?
final_df['Cuisines'].value_counts().head(10).reset_index(name='Restaurant Count')
```

```python
## Find whether Aggregate Rating is higher on average for:Restaurants with table booking or Restaurants with online delivery.
# Average rating by table booking
table_booking_rating = final_df.groupby('Has Table booking')['Aggregate rating'].mean().reset_index()

# Average rating by online delivery
online_delivery_rating = final_df.groupby('Has Online delivery')['Aggregate rating'].mean().reset_index()

print("Average rating by Table Booking:\n", table_booking_rating)
print("\nAverage rating by Online Delivery:\n", online_delivery_rating)
```

```python
## Do higher price restaurants get better ratings?
final_df.groupby('Price range')['Aggregate rating'].mean().reset_index().sort_values(by='Aggregate rating', ascending=False)
```

Yes, Higher price range has a higher rating.

```python
## Are highly-rated restaurants getting more votes?
final_df[['Aggregate rating','Votes']].corr()
```
